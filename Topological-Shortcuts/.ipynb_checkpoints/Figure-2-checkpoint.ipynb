```python
low_auroc = [0.5,0.5,0.5,0.5,0.5]
low_auprc = [0.008802060970373551,0.010020906928123423,0.009493446687920652,0.00903420090342009,0.00987831066571224]
high_auroc = [0.8499282285217536,0.8497072975709009,0.8438006352423153,0.8486056470443061,0.8427959647584465]
high_auprc = [0.8671682504444146,0.8663828427616982,0.8619140460869027,0.8669626708216118,0.8620929453150907]
```

```python
plt.figure(figsize=(8, 8),dpi=300)
categories = ['','']#['Low Degree', 'High Degree']
values1 = [np.mean(low_auroc), np.mean(high_auroc)]
values2 = [np.mean(low_auprc), np.mean(high_auprc)]

errors1 = [np.std(low_auroc), np.std(high_auroc)]
errors2 = [np.std(low_auprc), np.std(high_auprc)]

bar_width = 0.35

x = np.arange(len(categories))

plt.bar(x, values1, bar_width, yerr=errors1, label='AUROC',alpha=0.3)
plt.bar(x + bar_width, values2, bar_width, yerr=errors2, label='AUPRC', alpha=0.3)

# Adding labels, title, and legend
#plt.ylabel('Performance')
plt.xticks(x + bar_width/2, categories)
plt.legend(prop={'size': 20})

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```
